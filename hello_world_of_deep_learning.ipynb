```python
#import the libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
#Define the model
layer = keras.layers.Dense(units=1, input_shape=[1])
model = tf.keras.Sequential([layer])
# model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
#compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#provide the data
input = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
output = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
#train the neural network
model.fit(input, output, epochs=500)
```

```python
#Test the mode or predict the value
print(model.predict([10.0]))
```
